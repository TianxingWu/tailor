```python
import pandas as pd
```

```python
def show_as_dataframe(list_of_namedtuples, keys=None):
    if isinstance(list_of_namedtuples[0], list):
        if not keys:
            keys = range(len(list_of_namedtuples))
        return pd.concat([pd.DataFrame(tup_list) for tup_list in list_of_namedtuples], keys=keys)
    return pd.DataFrame(list_of_namedtuples)
```

# Build your own perturbation prompts

```python
from tailor.steps.perturb_prompt import PerturbPromptWithString
from tailor.common.perturbation_criteria import PerturbationCriteria
```

### Criteria

By default, perturbations are applied to all verbs in the sentence. Let's say we only want to apply the perturbation to verbs that include a relative clause.

```python
class RelativeClauses(PerturbationCriteria):
    def __call__(self, processed_sentence, *args, **kwargs):
        """ Extracts all verbs with an argument that includes a relative clause
        Returns:
            list[list[str]]: list of list of tags
                Each sub-list corresponds to a different verb
        """
        verbs = processed_sentence.verbs
        rel_idxes = [idx for idx, verb_dict in enumerate(verbs) \
                        if any(t.startswith("B-R") for t in verb_dict['tags'])]
        #if not rel_idxes: raise PerturbationError
        if not rel_idxes:
            return []
        tags_lst = [verbs[rel_idx]['tags'] for rel_idx in rel_idxes]
        return tags_lst
```

### Perturbation

```python
from tailor.common.perturb_function import PerturbStringFunction

class UntangleRelativeClauses(PerturbStringFunction):
    def __call__(
        self, prompt_meta, *args, description=None, **kwargs
    ):
        # delete context removes the relative pronoun (link to docs)
        perturb_str = (
            f"CONTEXT(DELETE_TEXT);NONCORE(ALL:DELETE)"
        )
        return Perturbation(
            perturb_str=perturb_str,
            perturb_meta=prompt_meta,
            name="untangle_relative_clauses",
            description=description,
        )
```

### Generate prompts

```python
new_prompts = PerturbPromptWithString().run(
    processed_sentences=processed_sentences,
    perturb_str_func=UntangleRelativeClauses(), 
    criteria_func=RelativeClauses(),
    intermediate_prompt_kwargs={"nblanks": 5}
)
```

```python
show_as_dataframe(new_prompts)
```

### Generate new sentences from prompts

```python
new_generations = GenerateFromPrompts().run(
    processed_sentences=processed_sentences,
    prompts=new_prompts,
    spacy_model=spacy_model,
    compute_perplexity=True,
)
```

```python
show_as_dataframe(new_generations)
```
